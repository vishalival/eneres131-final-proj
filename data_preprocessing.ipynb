### Full Data Pre-Processing Pipeline

```python
import pandas as pd
import zipfile
```

```python
zip_file_path = "final_data/biologicalresult2.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_bio_metadata = pd.read_csv(f, low_memory=False)

df_bio_metadata.head()
```

```python
valid_activity_ids = df_bio_metadata["ActivityIdentifier"].value_counts()[lambda x: x > 100].index

filtered_df_bio = df_bio_metadata[df_bio_metadata["ActivityIdentifier"].isin(valid_activity_ids)][[
    "OrganizationIdentifier", 
    "ActivityIdentifier",
    "ActivityStartDate", 
    "ActivityEndDate", 
    "CharacteristicName", 
    "ResultMeasureValue",
    "ResultMeasure/MeasureUnitCode", 
    "MeasureQualifierCode",
    "ResultStatusIdentifier",
    "ProjectIdentifier",
    "MonitoringLocationIdentifier"
    
]]

filtered_df_bio
```

```python
zip_file_path = "final_data/project2.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_project = pd.read_csv(f, low_memory=False)

df_project.head()
```

```python
zip_file_path = "final_data/station2.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_site = pd.read_csv(f, low_memory=False)

df_site.head()
```

```python
# Data Cleaning to drop Nans, and merging

filtered_df_bio = filtered_df_bio.dropna(subset=["ResultMeasureValue"])
filtered_df_bio_joined = filtered_df_bio.merge(df_project, how = "left", left_on = "ProjectIdentifier", right_on = "ProjectIdentifier")
filtered_df_bio_joined = filtered_df_bio_joined.merge(df_site, how = "left", left_on = "MonitoringLocationIdentifier", right_on = "MonitoringLocationIdentifier")
filtered_df_bio_joined = filtered_df_bio_joined.dropna(axis=1, how='all')
filtered_df_bio_joined['ResultMeasureValue'] = pd.to_numeric(filtered_df_bio_joined['ResultMeasureValue'], errors='coerce')
filtered_df_bio_joined.head()
```

```python
filtered_df_bio_joined.head()
```

```python
# A look at how many different unit measurements there are to measure a certain Characteristic!

unit_check = filtered_df_bio_joined.groupby("CharacteristicName")["ResultMeasure/MeasureUnitCode"].apply(lambda x: x.dropna().unique())

multiple_units = unit_check[unit_check.apply(len) > 1]

if multiple_units.empty:
    print("Each CharacteristicName has only one unique unit type of measurement.")
else:
    print("Some CharacteristicNames have multiple unit types of measurement:")
    for characteristic, units in multiple_units.items():
        print(f"- {characteristic}: {', '.join(units)}")
```

Each of the conversions, keeping and maintaining unit consistency.

```python
# ug/L to mg/kg case scenario
# Assumption: water has a density of 1 kg/L

row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ug/l"

# Dividing by 1000
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] / 1000
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# % to mg/kg case scenario
row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "%"

# Multiplying by 10,000
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * 10000
)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# ng/L to mg/kg case scenario
row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ng/l"

# Dividing by 1,000,000
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] / 1_000_000
)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# ft to m case scenario
row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ft"

# Multiplying by 0.3048
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * 0.3048
)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "m"
```

```python
# ft3/s to m3/s case scenario
row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ft3/s"

# Multiplying by 0.0283168
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * 0.0283168
)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "m3/sec"
```

```python
# mg/L as N to mg/L NH4 case scenario (if applicable)
row_indices_to_convert = filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l as N"

# Multiplying by 1.29 (example molecular weight conversion)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * 1.29
)
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/l NH4"
```

```python
# mg/kg to ug/kg only for certain edge cases. Using properly of *1000

row_indices_to_convert = (
    (filtered_df_bio_joined["CharacteristicName"].isin(["Fipronil Sulfone", "Fluoranthene", "Pyrene"])) &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/kg")
)

filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * 1000
)

filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "ug/kg"
```

```python
# mg/l to mg/kg only for certain edge cases. 

# assuming density = 1 for water

density = 1

row_indices_to_convert = (
    (filtered_df_bio_joined["CharacteristicName"].isin(["Calcium", "Magnesium", "Potassium", "Sodium"])) &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l")
)

filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] = (
    filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] * density
)

filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# List of CharacteristicNames to convert from ug/kg to mg/kg
characteristics_to_convert = [
    "Azoxystrobin", "Benzo[a]pyrene", "Clothianidin", "Dithiopyr", "Diuron",
    "Etoxazole", "Fipronil Sulfide", "Fluridone", "Hexazinone", "Methoxyfenozide",
    "Oxyfluorfen", "Pendimethalin", "Piperonyl butoxide", "Propiconazole",
    "Tin", "Uranium"
]

# Filter rows where conversion is needed
row_indices_to_convert = (
    (filtered_df_bio_joined["CharacteristicName"].isin(characteristics_to_convert)) &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ug/kg")
)

# Perform the conversion
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasureValue"] /= 1000
filtered_df_bio_joined.loc[row_indices_to_convert, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Convert mg/l NH4 to mg/l asNO3 for Nitrate
nitrate_indices = (filtered_df_bio_joined["CharacteristicName"] == "Nitrate") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l NH4"
)
filtered_df_bio_joined.loc[nitrate_indices, "ResultMeasureValue"] *= 4.43
filtered_df_bio_joined.loc[nitrate_indices, "ResultMeasure/MeasureUnitCode"] = "mg/l asNO3"
```

```python
# Convert mg/l NH4 to mg/l asNO2 for Nitrite
nitrite_indices = (filtered_df_bio_joined["CharacteristicName"] == "Nitrite") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l NH4"
)
filtered_df_bio_joined.loc[nitrite_indices, "ResultMeasureValue"] *= 4.43
filtered_df_bio_joined.loc[nitrite_indices, "ResultMeasure/MeasureUnitCode"] = "mg/l asNO2"
```

```python
# Convert % by wt to mg/kg for Organic Carbon
organic_carbon_indices = (filtered_df_bio_joined["CharacteristicName"] == "Organic carbon") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "% by wt"
)
filtered_df_bio_joined.loc[organic_carbon_indices, "ResultMeasureValue"] *= 10000
filtered_df_bio_joined.loc[organic_carbon_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Convert mg/l as PO4 to mg/l as P for Orthophosphate
orthophosphate_indices = (filtered_df_bio_joined["CharacteristicName"] == "Orthophosphate") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l asPO4"
)
filtered_df_bio_joined.loc[orthophosphate_indices, "ResultMeasureValue"] *= 0.326
filtered_df_bio_joined.loc[orthophosphate_indices, "ResultMeasure/MeasureUnitCode"] = "mg/l as P"
```

```python
# Convert % saturation to mg/l for Oxygen
oxygen_indices = (filtered_df_bio_joined["CharacteristicName"] == "Oxygen") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "% saturatn"
)
# Example multiplier, replace with correct value if known
filtered_df_bio_joined.loc[oxygen_indices, "ResultMeasureValue"] *= 0.1
filtered_df_bio_joined.loc[oxygen_indices, "ResultMeasure/MeasureUnitCode"] = "mg/l"
```

```python
# Standardize mg/kg as P to mg/kg for Phosphorus
phosphorus_indices = (filtered_df_bio_joined["CharacteristicName"] == "Phosphorus") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/kg as P"
)
filtered_df_bio_joined.loc[phosphorus_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Convert mg/l to mg/kg for SSC
ssc_indices = (filtered_df_bio_joined["CharacteristicName"] == "Suspended Sediment Concentration (SSC)") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l"
)
# Assuming density of 1 kg/L
filtered_df_bio_joined.loc[ssc_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Convert tons/ac ft to mg/l for Total Dissolved Solids
tds_indices = (filtered_df_bio_joined["CharacteristicName"] == "Total dissolved solids") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "tons/ac ft"
)
# Conversion factor: 1 ton/ac ft = 73,558.8 mg/l
filtered_df_bio_joined.loc[tds_indices, "ResultMeasureValue"] *= 73558.8
filtered_df_bio_joined.loc[tds_indices, "ResultMeasure/MeasureUnitCode"] = "mg/l"
```

```python
# Convert units/cm to L/mgDOC*m for UV 254
uv_indices = (filtered_df_bio_joined["CharacteristicName"] == "UV 254") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "units/cm"
)
filtered_df_bio_joined.loc[uv_indices, "ResultMeasure/MeasureUnitCode"] = "L/mgDOC*m"
```

```python
# Convert mg/l to mg/kg for "Organic carbon" minority case
final_oc_indices = (filtered_df_bio_joined["CharacteristicName"] == "Organic carbon") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l"
)

filtered_df_bio_joined.loc[final_oc_indices, "ResultMeasureValue"] /= 1000
filtered_df_bio_joined.loc[final_oc_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Convert mg/l as P to mg/kg for "Phosphorus" minority case
phosphorus_indices = (filtered_df_bio_joined["CharacteristicName"] == "Phosphorus") & (
    filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "mg/l as P"
)

# Perform the conversion
filtered_df_bio_joined.loc[phosphorus_indices, "ResultMeasureValue"] /= 1000
filtered_df_bio_joined.loc[phosphorus_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Methane and Nitrogen: Convert cm3/g STP to cm3/g @STP
methane_nitrogen_indices = (
    (filtered_df_bio_joined["CharacteristicName"].isin(["Methane", "Nitrogen"])) &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "cm3/g STP")
)

# Update the unit
filtered_df_bio_joined.loc[methane_nitrogen_indices, "ResultMeasure/MeasureUnitCode"] = "cm3/g @STP"
```

```python
# Organic carbon: Convert g/kg to mg/kg
organic_carbon_indices = (
    (filtered_df_bio_joined["CharacteristicName"] == "Organic carbon") &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "g/kg")
)

# Perform the conversion
filtered_df_bio_joined.loc[organic_carbon_indices, "ResultMeasureValue"] *= 1000
filtered_df_bio_joined.loc[organic_carbon_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# Turbidity: Convert FNU to NTU
turbidity_indices = (
    (filtered_df_bio_joined["CharacteristicName"] == "Turbidity") &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "FNU")
)

# Update the unit (values remain unchanged)
filtered_df_bio_joined.loc[turbidity_indices, "ResultMeasure/MeasureUnitCode"] = "NTU"
```

```python
# Uranium: Convert pCi/L to mg/kg
uranium_indices = (
    (filtered_df_bio_joined["CharacteristicName"] == "Uranium") &
    (filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "pCi/L")
)

# Perform the conversion (replace the factor with the correct value if needed)
filtered_df_bio_joined.loc[uranium_indices, "ResultMeasureValue"] *= 0.000000034
filtered_df_bio_joined.loc[uranium_indices, "ResultMeasure/MeasureUnitCode"] = "mg/kg"
```

```python
# FINAL CHECK

unit_check = filtered_df_bio_joined.groupby("CharacteristicName")["ResultMeasure/MeasureUnitCode"].apply(lambda x: x.dropna().unique())

multiple_units = unit_check[unit_check.apply(len) > 1]

if multiple_units.empty:
    print("Each CharacteristicName has only one unique unit type of measurement.")
else:
    print("Some CharacteristicNames have multiple unit types of measurement:")
    for characteristic, units in multiple_units.items():
        print(f"- {characteristic}: {', '.join(units)}")
```
