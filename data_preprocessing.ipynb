### Full Data Pre-Processing Pipeline

```python
import pandas as pd
import zipfile
```

```python
zip_file_path = "final_data/biologicalresult.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_bio_metadata = pd.read_csv(f, low_memory=False)

df_bio_metadata.head()
```

```python
valid_activity_ids = df_bio_metadata["ActivityIdentifier"].value_counts()[lambda x: x > 100].index

filtered_df_bio = df_bio_metadata[df_bio_metadata["ActivityIdentifier"].isin(valid_activity_ids)][[
    "OrganizationIdentifier", 
    "ActivityIdentifier",
    "ActivityStartDate", 
    "ActivityEndDate", 
    "CharacteristicName", 
    "ResultMeasureValue",
    "ResultMeasure/MeasureUnitCode", 
    "MeasureQualifierCode",
    "ResultStatusIdentifier",
    "ProjectIdentifier",
    "MonitoringLocationIdentifier"
    
]]

filtered_df_bio
```

```python
zip_file_path = "final_data/project.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_project = pd.read_csv(f, low_memory=False)

df_project.head()
```

```python
zip_file_path = "final_data/station.zip"

with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    
    csv_file_name = [name for name in file_names if name.endswith('.csv')][0]
    
    with z.open(csv_file_name) as f:
        df_site = pd.read_csv(f, low_memory=False)

df_site.head()
```

```python
filtered_df_bio_joined = filtered_df_bio.merge(df_project, how = "left", left_on = "ProjectIdentifier", right_on = "ProjectIdentifier")
filtered_df_bio_joined = filtered_df_bio_joined.merge(df_site, how = "left", left_on = "MonitoringLocationIdentifier", right_on = "MonitoringLocationIdentifier")
filtered_df_bio_joined = filtered_df_bio_joined.dropna(axis=1, how='all')
filtered_df_bio_joined['ResultMeasureValue'] = pd.to_numeric(filtered_df_bio_joined['ResultMeasureValue'], errors='coerce')
filtered_df_bio_joined
```

```python
# Define a conversion function for ug/l to mg/l
def convert_ug_l_to_mg_l(row):
    if row["ResultMeasure/MeasureUnitCode"] == "ug/l" and not pd.isna(row["ResultMeasureValue"]):
        # Convert value and update unit
        return row["ResultMeasureValue"] / 1000, "mg/l"
    # If no conversion needed, return original values
    return row["ResultMeasureValue"], row["ResultMeasure/MeasureUnitCode"]

# Apply the conversion to the DataFrame
filtered_df_bio_joined[["ResultMeasureValue", "ResultMeasure/MeasureUnitCode"]] = filtered_df_bio_joined.apply(
    lambda row: pd.Series(convert_ug_l_to_mg_l(row)), axis=1
)

# Verify if any ug/l remains in the unit column
remaining_ug_l = filtered_df_bio_joined[filtered_df_bio_joined["ResultMeasure/MeasureUnitCode"] == "ug/l"]
if remaining_ug_l.empty:
    print("All ug/l values have been successfully converted to mg/l.")
else:
    print("Some ug/l values remain unconverted:")
    print(remaining_ug_l)
```

```python
# Group by CharacteristicName and find the unique MeasureUnitCode for each
unit_check = filtered_df_bio_joined.groupby("CharacteristicName")["ResultMeasure/MeasureUnitCode"].nunique()

# Filter for CharacteristicName values with more than one unique MeasureUnitCode
multiple_units = unit_check[unit_check > 1]

if multiple_units.empty:
    print("Each CharacteristicName has only one unique unit type of measurement.")
else:
    print("Some CharacteristicNames have multiple unit types of measurement:")
    print(multiple_units)
```
